```python
def new_column(total_account_info):
  if total_account_info["Distance to Closest Branch (mi)"] == total_account_info['Distance to Branch A (mi)']:
    return "Branch A"
  elif total_account_info["Distance to Closest Branch (mi)"] == total_account_info['Distance to Branch B (mi)']:
    return "Branch B"
  elif total_account_info["Distance to Closest Branch (mi)"] == total_account_info['Distance to Branch C (mi)']:
    return "Branch C"
  elif total_account_info["Distance to Closest Branch (mi)"] == total_account_info['Distance to Branch D (mi)']:
    return "Branch D"
  else:
    print("ERROR")
```

#Phase 1- *Retention Model*

```python
def secondbranch(total_account_info):
  if total_account_info["Distance to Closest Branch (mi)"] == total_account_info["Distance to Branch A (mi)"]:
    return total_account_info[["Distance to Branch B (mi)","Distance to Branch C (mi)","Distance to Branch D (mi)"]].min() - total_account_info["Distance to Closest Branch (mi)"]
  elif total_account_info["Distance to Closest Branch (mi)"] == total_account_info["Distance to Branch B (mi)"]:
    return total_account_info[["Distance to Branch A (mi)","Distance to Branch C (mi)","Distance to Branch D (mi)"]].min() - total_account_info["Distance to Closest Branch (mi)"]
  elif total_account_info["Distance to Closest Branch (mi)"] == total_account_info["Distance to Branch C (mi)"]:
    return total_account_info[["Distance to Branch A (mi)","Distance to Branch B (mi)","Distance to Branch D (mi)"]].min() - total_account_info["Distance to Closest Branch (mi)"]
  elif total_account_info["Distance to Closest Branch (mi)"] == total_account_info["Distance to Branch D (mi)"]:
    return total_account_info[["Distance to Branch A (mi)","Distance to Branch B (mi)","Distance to Branch C (mi)"]].min() - total_account_info["Distance to Closest Branch (mi)"]
  else:
    print("Error")
```

```python
#Run cell one time.
total_account_info["2nd Closest Branch"]= total_account_info.apply(secondbranch, axis=1)
```

```python
def secbranch_mi(total_account_info):
  if total_account_info['Closest Branch']== "Branch D":
    return total_account_info[['Distance to Branch A (mi)','Distance to Branch B (mi)', 'Distance to Branch C (mi)']].min()
  elif total_account_info['Closest Branch']== "Branch C":
    return total_account_info[['Distance to Branch A (mi)','Distance to Branch B (mi)', 'Distance to Branch D (mi)']].min()
  elif total_account_info['Closest Branch']== "Branch B":
    return total_account_info[['Distance to Branch A (mi)','Distance to Branch C (mi)', 'Distance to Branch D (mi)']].min()
  elif total_account_info['Closest Branch']== "Branch A":
    return total_account_info[['Distance to Branch B (mi)','Distance to Branch C (mi)', 'Distance to Branch D (mi)']].min()
  else:
    print("Error")
total_account_info["2nd Closest Branch"]= total_account_info.apply(secbranch_mi, axis=1)
total_account_info
def secbranch_letter(total_account_info):
  if total_account_info['2nd Closest Branch']== total_account_info['Distance to Branch A (mi)']:
    return 'Branch A'
  elif total_account_info['2nd Closest Branch']== total_account_info['Distance to Branch B (mi)']:
    return 'Branch B'
  elif total_account_info['2nd Closest Branch']== total_account_info['Distance to Branch C (mi)']:
    return 'Branch C'
  elif total_account_info['2nd Closest Branch']== total_account_info['Distance to Branch D (mi)']:
    return 'Branch D'
  else:
    print("ERROR")
total_account_info['next closest']= total_account_info.apply(secbranch_letter,axis=1)
total_account_info
total_account_info_2= total_account_info.drop('2nd Closest Branch',axis=1)
total_account_info_2.rename(columns= {"next closest": "2nd Closest Branch","Distance to Closest Branch (mi)":"Dist. to Closest Branch (mi)"},inplace=True)
total_account_info_2
```
